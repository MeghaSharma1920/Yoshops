### Import Required Libraries:

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

# MOBILE

### Base URL and URL:

```python
Base_url = "https://www.flipkart.com/search?q=mobile%20&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off"
```

```python
url = "https://www.flipkart.com/search?q=mobile%20&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page="
```

### Status_Code:

```python
req=requests.get(Base_url)
```

```python
req
```

### Creating Empty Lists:

```python
n1=[]
pr1=[]
r1=[]
d1=[]
```

```python
for i in range(1,41):
    url=Base_url+str(i)
    req=requests.get(url)
    soup=BeautifulSoup(req.text,"html.parser")
    name=soup.find_all("div",class_="_4rR01T")
    processor=soup.find_all("li",class_="rgWa7D")
    rating=soup.find_all("span",class_="_2_R_DZ")
    off=soup.find_all("div",class_="_3Ay6Sb")
    for n,p1,r,o in zip(name,processor,rating,off):
        n1.append(n.text)
        pr1.append(p1.text)
        r1.append(r.text)
        d1.append(o.text)
        
        

    
```

### Print All data to check it is appended or not:

```python
print(n1,pr1,r1,d1)
```

### Data Frame:

```python
df=pd.DataFrame({"Mobile_name":n1,"Specifics":pr1,"Rating & Reviews":r1," Discount":d1})
```

```python
df
```

### DataFrame to .CSV:

```python
df.to_csv("Mobiles data.csv")
```

### Status_Code:

```python
req=requests.get(Base_url)
req
```

### Creating Empty Lists:

```python
a1 =[]
b1 =[]
c1 =[]
d1 =[]
e1 =[]
f1 =[]
g1 =[]
h1 =[]
```

```python
for i in range(1,41):
    url=Base_url+str(i)
    req=requests.get(url)
    soup=BeautifulSoup(req.text,"html.parser")
    name=soup.find_all("div",class_="_4rR01T")
    rating=soup.find_all("div",class_="_3LWZlK")
    No_of_rating=soup.find_all("span",class_="_2_R_DZ")
    No_of_reviews=soup.find_all("span",class_="_13vcmD")
    Specific=soup.find_all("div",class_="fMghEO")
    Discounted_Price=soup.find_all("div",class_="_30jeq3 _1_WHN1")
    Original_Price=soup.find_all("div",class_="_3I9_wc _27UcVY")
    Discount=soup.find_all("div",class_="_3Ay6Sb")
    for a,b,c,d,e,f,g,h in zip(name,rating,No_of_rating,No_of_reviews,Specific,Discounted_Price,Original_Price,Discount):
        a1.append(a.text)
        b1.append(b.text)
        c1.append(c.text)
        d1.append(d.text)
        e1.append(e.text)
        f1.append(f.text)
        g1.append(g.text)
        h1.append(h.text)
        
```

### Print All data to check it is appended or not:

```python
print(a1,b1,c1,d1,e1,f1,g1,h1)
```

### Data Frame:

```python
df1=pd.DataFrame({"Laptop_name":a1,"rating":b1,"No_of_rating":c1,"Specific":e1,"Discounted_Price":f1,"Original_Price":g1,"Discount":h1})
df1
```
